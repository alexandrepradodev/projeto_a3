O próximo passo é transformar as colunas que são os anos em uma lista, pegando apenas os 4 primeiros números e multiplicar por 195 que é a quantidade de países.

```python
lista_dos_anos = list(dados_obesidade_sujo.columns)
del lista_dos_anos[0]
lista_dos_anos = [x[0:4] for x in lista_dos_anos]
lista_dos_anos_multiplicada = lista_dos_anos * 195
# lista_dos_anos_multiplicada.sort()
lista_dos_anos_multiplicada
```

```python
dados_dict = {'Country': pd.Series(paises_multiplicado),
              'Year': pd.Series(lista_dos_anos_multiplicada)}
```

Agora iremos adicionar a coluna dos sexos e para isso cada um dos sexos será repetido 8190 vezes, sendo que deve seguir a ordem Both Sexes, Male, Female

```python
lista_sexos = []

for i in range(8190):
    lista_sexos.append('Both Sexes')
    lista_sexos.append('Male')
    lista_sexos.append('Female')
```

```python
dados_dict = {'Country': pd.Series(paises_multiplicado),
              'Year': pd.Series(lista_dos_anos_multiplicada),
              'Sex': pd.Series(lista_sexos)}
```
